```python
import numpy as np
from scipy import integrate
from scipy.integrate import solve_ivp
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

```python
# Є набір даних speed, який представляє значення швидкості для деякого транспортного засобу в певний момент спостереження. 
# Очевидно, що дані мають дискретний вигляд. Відомо, що спостереження відбувалися з періодом в одну годину.
speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]
```

```python
# Приймемо, що в нульовій координаті у нас швидкість 25 км/год, відповідно до набору даних speed. 
# Помістіть у змінну time - вектор часу, створений за допомогою np.linspace (всього 12 спостережень, від 0 до 11 годин)
time = np.linspace(0, 11, num=12)
time
```

```python
# Виконайте виведення графіка — значення точок швидкості (plot або scatter). 
# Задайте розмір області, що відображається (0, 11) і (0, 130). Задайте відображення сітки
plt.plot(time, speed)
plt.grid(True)
plt.xticks(range(0, 12))
plt.show()
```

```python
# Виконайте інтерполяцію за допомогою interp1d(kind='cubic') та отримайте функцію на 10000 значень. 
# Побудуйте безперервний графік отриманої функції.
time_interp = np.linspace(0, 11, num=10000)
speed_interp = interp1d(time, speed, kind='cubic')
plt.plot(time_interp, speed_interp(time_interp))
plt.grid(True)
plt.xticks(range(0, 12))
plt.show()
```

```python
# Відповідно до фізичного змісту першої похідної, похідна функції у точці є миттєва швидкість точки,
# Обчисліть інтеграл для отриманої інтерполяційної функції на проміжку [0, 11]
distance_cubic = integrate.quad(speed_interp, 0, 11)
print(distance_cubic)
```

```python
# Виконайте аналогічно for kind='quadratic'
time_interp = np.linspace(0, 11, num=10000)
speed_interp = interp1d(time, speed, kind='quadratic')
plt.plot(time_interp, speed_interp(time_interp))
plt.grid(True)
plt.xticks(range(0, 12))
plt.show()
```

```python
distance_quadratic = integrate.quad(speed_interp, 0, 11)
print(distance_quadratic)
```
