```python
import numpy as np
print(f"numpy version: {np.__version__}")
import pandas as pd
print(f"pandas version: {pd.__version__}")
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://textbook.edu.goit.global/python/data-science-xf8b2g-pwa/v1/uk/docs/module-03/hw'
df = pd.read_html(url)[0]
df
```

```python
# build regplot that shows the profitability of one stock action on another
sns.regplot(x="Доходность A, %", y="Доходность B, %", data = df, label = "A vs. B", color='darkblue')
sns.regplot(x="Доходность A, %", y="Доходность C, %", data = df, label = "A vs. C", color='green')
sns.regplot(x="Доходность B, %", y="Доходность C, %", data = df, label = "B vs. C", color='magenta')
plt.legend()
plt.show()
```

```python
# build regplot that shows the profitability of one stock action on another
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

sns.regplot(x="Доходность A, %", y="Доходность B, %", data = df, ax=axes[0], color='darkblue').set(title = "A vs. B")
sns.regplot(x="Доходность A, %", y="Доходность C, %", data = df, ax=axes[1], color='green').set(title = "A vs. C")
sns.regplot(x="Доходность B, %", y="Доходность C, %", data = df, ax=axes[2], color='magenta').set(title = "B vs. C")
# plt.legend()
plt.show()
```

```python
# get the correlation beatween stocks
df.corr()
```

```python
# create a correlation matrix that measures the linear relationships between the variables
correlation_matrix = df.corr().round(2)
# plot using seaborn lib
plt.figure(figsize = (16,5))
plt.title('Correlation matrix')
sns.heatmap(data=correlation_matrix, annot=True) # annot = True to print the values inside the square
```

From the correlation matrix one sees that the largest correlation is between B and C,
between C and A we see even anticorrelation, the smalles correlation is between B and A.
