# Sierpinski Triangle

```python
import matplotlib.pyplot as plt
from random import randint

x1 = -20
y1 = 0
x2 = 20
y2 = 0
x3 = 0
y3 = 20
x0 = 0
y0 = 0
max_range = 200000

x_coords = []
y_coords = []

for i in range(0,max_range):
    r = randint(0,4)
    if r == 1:
        x0 = (x0 + x1) / 2
        y0 = (y0 + y1) / 2
    if r == 2:
        x0 = (x0 + x2) / 2
        y0 = (y0 + y2) / 2
    if r == 3:
        x0 = (x0 + x3) / 2
        y0 = (y0 + y3) / 2
    x_coords.append(x0)
    y_coords.append(y0)

plt.title('Sierpinski Triangle')
fig = plt.scatter(x_coords, y_coords, s=0.02, color='green')
plt.rcParams["figure.figsize"] = (10,10)
plt.show()
```
